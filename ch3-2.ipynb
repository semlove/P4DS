```python
import matplotlib.pyplot as plt
%matplotlib inline

values = [10, 20, 30, 15, 5]
colors = ['b', 'g', 'r', 'c', 'm']
labels = ['A', 'B', 'C', 'D', 'E']
explode = (0, 0, 0.4, 0, 0)

plt.pie(values, colors=colors, labels=labels,
       explode=explode, autopct='%1.1f%%',
        counterclock=False, shadow=True)
plt.title('Percent of Values')

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

values = [10, 20, 30, 15, 5]
widths = [0.7, 0.7, 0.8, 0.7, 0.7]
colors = ['b', 'b', 'r', 'b', 'b']

plt.bar(range(0, 5), values, width=widths, color=colors, align='center')

plt.title('Number of Values')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = [1, 2, 3, 4, 5, 6]
y1 = [10, 20, 25, 30, 35, 40]
y2 = [25, 14, 26, 15, 15, 20]
labels = ["A", "B", "C", "D", "E", "F"]

plt.bar(x, y1, tick_label=labels)
plt.bar(x, y2, bottom=y1)

plt.legend(("y1", "y2"))
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")
g = sns.catplot(x="who", y="survived", col="class",
        data=titanic, kind="bar", ci=None, aspect=1)

g.set_axis_labels("", "Survival Rate")
g.set_xticklabels(["Men", "Women", "Children"])
g.set_titles("{col_name} {col_var}")

#---show plot---
plt.show()
```
