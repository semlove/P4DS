```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = 20*np.random.randn(10000)

plt.hist(x, 10, range=(-60, 60), histtype='stepfilled',
        align='mid', color='g', label="Test Data")
plt.legend()
plt.title('Normal Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = 20*np.random.randn(10000)
plt.hist(x, bins=100, normed=True, cumulative=True) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Plot between -3 and 3with 1steps.
x_axis = np.arange(-3, 3, 0.01)
# Mean = 0, SD = 3.
plt.plot(x_axis, norm.pdf(x_axis,0,1))
plt.show()
```
